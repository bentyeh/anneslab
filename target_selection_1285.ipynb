```python
# Import all packages
import numpy as np
import pandas as pd
```

```python
## parameters
percent_thresh_high = 90 # set <= 0 to ignore this parameter
percent_thresh_low = 100 # set >= 100 to ignore this parameter
diff_percent_thresh = 30
```

```python
## Read in data
df1 = pd.read_csv('Annes(2018)_Kinome_1285.csv')
df2 = pd.read_csv('Annes(2018)_Kinome500_1285.csv')
```

```python
# merge 100 nM and 500 nM datasets
merged = pd.merge(df1, df2, how='inner', on=['Compound Name','DiscoveRx Gene Symbol', 'Entrez Gene Symbol'], suffixes=(" (100 nM)", " (500 nM)"))
merged.drop(['Compound Concentration (nM) (100 nM)', 'Compound Concentration (nM) (500 nM)'], axis=1, inplace=True)
merged['diff'] = merged['Percent Control (100 nM)'] - merged['Percent Control (500 nM)']
# merged.to_csv('Annes(2018)_Kinome_1285_comp.csv', index=False)
merged
```

```python
# get top results
merged.sort_values(by='diff', ascending=False, inplace=True)
result = set(
    merged.loc[(merged['diff'] >= diff_percent_thresh) &
               (merged['Percent Control (500 nM)'] <= percent_thresh_low) &
               (merged['Percent Control (100 nM)'] >= percent_thresh_high), 'Entrez Gene Symbol'])
result
```
